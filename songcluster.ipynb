```python
features = song_data[['danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness','liveness', 'valence', 'tempo']]
```

```python
features.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import RobustScaler
```

```python
song_scaler = RobustScaler()
```

```python
scaled_features = song_scaler.fit_transform(features)
print("scaled_features:\n",scaled_features)
plot_scaled_features =pd.DataFrame(scaled_features,columns =features.columns)
plot_scaled_features
hist_sacled_features = plot_scaled_features.hist(figsize=(12,10),bins=20, color ='blue',edgecolor ='black')
```

```python
from sklearn import neighbors
from sklearn.neighbors import NearestNeighbors
from sklearn.cluster import DBSCAN
```

```python
# finding the optimal number of epsilon and min samples for DBSCAN using the K-distance graph

k = 8 #(D*2)
knbrs = neighbors.NearestNeighbors(n_neighbors = k,algorithm ='ball_tree').fit(scaled_features)
distances, indices = knbrs.kneighbors(scaled_features)

print("distances:\n",distances)
print("indices:\n",indices)

```

```python
# mapping and including the indices and distances back into the original dataset

act_indices = song_data.index.values
mapped_indices = act_indices[indices]
```

```python
dist_cols = [f'dist_neighbor_{i}' for i in range(k)]
song_distances = pd.DataFrame(distances, columns=dist_cols, index= song_data.index)

idx_cols = [f'idx_neighbor_{i}' for i in range(k)]
song_indices = pd.DataFrame(indices, columns=idx_cols, index=song_data.index)
```

```python
song_data_updated = pd.concat([song_data, song_distances, song_indices], axis=1)
```

```python
sorted_distances = np.sort(distances,axis =0)
plt.plot(sorted_distances)
plt.title("K-Distance Plot to Find Epsilon")
plt.xlabel("Songs sorted by distance")
plt.ylabel(f"{k}-NN Distance")
plt.show()

# approximately about 1500 samples are within the distance of 1
```

```python
from sklearn.cluster import DBSCAN
```

```python
# initializing the DBSCAN algorithm with the optimal values of epsilon and min_samples
dbscan  = DBSCAN(eps = 0.65 , min_samples = 8, metric ='euclidean', algorithm ='auto')
```

```python
# fitting the standardized dataframe of features into the DBSCAN algorithm.

song_cohort_labels = dbscan.fit(scaled_features)
```

```python
type(song_cohort_labels)
```

```python
song_data_updated['cohort_index'] = song_cohort_labels.labels_
```

```python
cohort_features = [ 'danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness','liveness', 'valence', 'tempo']
```

```python
cohort_profiles = song_data_updated.groupby('cohort_index')[cohort_features].mean()
```

```python
print(cohort_profiles)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
sns.heatmap(cohort_profiles, annot=True, fmt=".2f", cmap="YlGnBu", cbar=True, linewidths=0.5)
plt.title('Cohort Feature Mean Heatmap', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Audio Features')
plt.ylabel('Cohort ID')
plt.tight_layout()
plt.show()
```

```python
# as the cluster is a non linear clsuter, wee use a t distribution method to project the data inti 2 D space for visulaization of teh clusters.
from matplotlib import colors, markers
from sklearn.manifold import TSNE

# 1. Project to 2D using t-SNE
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
X_tsne = tsne.fit_transform(scaled_features)

# 2. Plot t-SNE coordinates
plt.figure(figsize=(9, 6))

palette = {-1: 'gray', 0: 'tab:blue', 1: 'tab:orange', 2: 'tab:green', 3: 'tab:red'}
marker_map = {-1: 'x', 0: 'o', 1: 's', 2: '^', 3: 'D'}

for cohort in sorted(song_data_updated['cohort_index'].unique()):
    mask = (song_data_updated['cohort_index'] == cohort)
    lbl = f"Cohort {cohort}" if cohort != -1 else "Noise (-1)"
    alpha = 0.4 if cohort == -1 else 0.7
    
    plt.scatter(
        X_tsne[mask, 0], X_tsne[mask, 1], 
        c=palette.get(cohort, 'purple'),
        marker=marker_map.get(cohort, 'o'), 
        label=lbl, 
        alpha=alpha, 
        edgecolors='k' if cohort != -1 else 'none', 
        linewidths=0.5
    )

plt.title('2D t-SNE Projection of DBSCAN Song Cohorts', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
plt.legend(title='Cohorts', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
